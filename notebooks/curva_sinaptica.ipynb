```python
import matplotlib.pyplot as plt
import numpy as np

# Variáveis
voltage = np.linspace(-2, 3, 500)  # Intervalo de voltagem pós-sináptica

# Função para variação da força sináptica
def plasticity_curve(voltage, threshold=0):
    return -np.exp(-2 * (voltage - threshold)) + np.exp(2 * (voltage - threshold))

# Calcular a variação da força sináptica (Δw) com base na voltagem
delta_w = plasticity_curve(voltage)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(voltage, delta_w, label=r'$\Delta w$', color='black', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.text(0, -0.5, 'Limiar LTP', ha='right', va='center', fontsize=10, color='black')
plt.xlabel("Voltagem Pós-sináptica")
plt.ylabel("Variação da Força Sináptica (Δw)")
plt.title("Curva de Plasticidade Dependente da Voltagem")
plt.grid(True)
plt.legend()
plt.show()
```
